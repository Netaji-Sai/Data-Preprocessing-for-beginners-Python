### **Handling NAN Missing Values**


*   Delete row/column
*   imputing values filling with a specific values
*if 70-80% is null in particular column ---- delete column




````python
import numpy as np
f1 = pd.DataFrame({"Name":["Netaji","Preeti",np.nan,"AAyush","Venky","Phani"],"Age":[np.nan,29,np.nan,19,24,np.nan],"City":["SKLM","NLR",np.nan,np.nan,"BZA","GTR"]})
f1
````

````python
f1.isnull()
````

````python
f1.isnull().any()
````

````python
f1.isnull().sum()
````

````python
#Deleting the rows conatining null value
#rows = axis=0
#columns = axis=1
f1.dropna(axis=0) #inplace=True if we want to commit
````

````python
# imputing with mean,median,mode
m = f1['Age'].mean() #mean
f1['Age'] = f1['Age'].fillna(m)
f1
````

````python
f1 = pd.DataFrame({"Name":["Netaji","Preeti",np.nan,"AAyush","Venky","Phani"],"Age":[np.nan,29,np.nan,19,18,np.nan],"City":["SKLM","NLR",np.nan,np.nan,"BZA","BZA"]})
med = f1['Age'].median() #meadian
f1['Age'] = f1['Age'].fillna(med)
f1
````

````python
f1 = pd.DataFrame({"Name":["Netaji","Preeti",np.nan,"Phani","Venky","Phani"],"Age":[np.nan,29,21,19,18,np.nan],"City":["SKLM","SKLM",np.nan,np.nan,"BZA","GTR"]})
f1['City']=f1['City'].fillna(f1['City'].mode()[0]) #mode generally stores as series as there can be 2 values have same no of repeats
f1['Name']=f1['Name'].fillna(f1['Name'].mode()[0])
med = f1['Age'].median() #meadian
f1['Age'] = f1['Age'].fillna(med)
f1
````

````python
f1.City.value_counts()
````

````python
group = f1.groupby("City")
print(group)
print(group.mean())
````
