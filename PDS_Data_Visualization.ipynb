### Analysis are of 3 types :
1.   UniVarient -> single variable(can be categoriacal->c/nummerical->n) at a time(Histogram, Barplot, pie charts, box plots)

2.   BiVarient -> b/w 2 variables
*   n vs n -> scatterplot(change in one variable wrt other variable)
*   n vs c -> ANOVA(analysis of varients)
*   c vs c -> Chisquared Error Test

3.   MultiVarient -> many variables
* Clustering -> segregate into groups
* PCA(Principal Component Analysis)
* Multi Correspondence Analysis - Surveys

### **Subplots**

```python
x = np.linspace(1,15,10)
y = x**2
plt.subplot(2,2,1)
plt.plot(x,y,'b--',ms=7)
plt.subplot(2,2,2)
plt.plot(x,y,'r*',mec='b')
plt.subplot(2,2,3)
plt.plot(x,y,'r-o')
plt.subplot(2,2,4)
plt.plot(x,y,'g-<')
```

###Box Plot -> detect the outliers
Outlier -> data point which is not usual from other data in same feature

```python
sns.boxplot(x = 'day',y='total_bill',data=df)
```

* Middle Line -> Meadian
* Above Middle -> Upper Quartile(Quartile->list of data grouped)
* Below Middle -> Lower Quartile
* Up Line -> Max Data
* Down Line -> Min Data
* Dots -> Outliers

Understanding About Quartiles :
Let Age = [10,20,30,40,50,60,80,200]
* 1st half =[10,20,30,40]

  Med -> (20+30)/2 = 25 ----> Q1(Lower)
*  Meadian -> (40+50)/2 = 45 ----> Q2(Middle)
* 2nd Half =[50,60,80,200]

   med -> (60+80)/2 = 70 ---->Q3(Upper)

* iqr(Inter Quartile Range) = Q3 - Q1
* If Q2 towards Q1 then Positively Skewed
* If Q2 towards Q3 then Negatively Skewed
